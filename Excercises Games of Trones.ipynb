```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/w/index.php' + \
      '?title=List_of_Game_of_Thrones_episodes&oldid=802553687'
```

```python
r = requests.get(url)
html=r.text
html_soup=BeautifulSoup(html,'html.parser')
```

```python
#extract all links
links= html_soup.find_all('a')
print(links)
```

```python
hrefs=[link.get('href') for link in links]
for item in hrefs:
    print(('{}').format(item))
```

```python
#extract images
images=html_soup('img')
for imgx in images:
    print(imgx)
    print('----------------------------------' )
    print(imgx.attrs)
    print('----------------------------------' )
    print(imgx['src'])
```

```python
#extract ratings table
final=[]
ep_tables=html_soup.find_all('table',class_='wikitable')
firstTable=ep_tables[0]
headers=[]
for header in firstTable.find('tr').find_all('th'):
    if header.text == 'Originally aired':
        headers.append('Originally aired - First aired')
        headers.append('Originally aired - Last aired')
    else:
        headers.append(header.text)
for row in firstTable.find_all('tr')[2:]:
    values=[]
    for col in row.find_all(['th','td']) :
        if not col.text=='':
            values.append(col.text)
    if values:
        dict_table={headers[i]:values[i] for i in range(len(values))}
        final.append(dict_table)
```

```python
df=pd.DataFrame(final)
df
```
